```python
import requests
import re
OVERPASS_URL = "http://localhost:8080/api/interpreter"
OVERPASS_URL = "https://overpass-api.de/api/interpreter"


PLACES = [
    "Mama's Too",
    "Lucali",
    "Luigi's",
    "L'Industrie Pizza",
    "F&F Pizzeria",
    "Stretch Pizza",
    "Ops",
    "Joe's Pizza",
    "Grand Street Pizza",
    "Una Pizza Napoletana",
    "Patsy's Pizzeria",
]

blocks = [f'   node(area.nyc)["name"~"{re.escape(n)}$"];' for n in PLACES]
blocks = "\n".join(blocks)

    


query = f"""
[out:json][timeout:180];

area["name"="United States"]["boundary"="administrative"]["admin_level"="2"]->.us;


area(area.us)
  ["name"="New York City"]
  ["boundary"="administrative"]
  ["admin_level"="5"]
  ->.nyc;

({blocks})

out;
"""
print(query)
response = requests.post(OVERPASS_URL, data=query.encode(
    "utf-8"), headers={"Content-Type": "application/x-www-form-urlencoded"})
response.raise_for_status()
data = response.json()

data
```
